```python
import os
import random
import telepot
from telepot.loop import MessageLoop
```

```python
def handle(msg):
  chat_id = msg['chat']['id']
  command = msg['text']
  if command == '/image':
    bot.sendPhoto(chat_id, 'https://picsum.photos/200/300/?random')
```

```python
BOT_TOKEN = 
bot = telepot.Bot(BOT_TOKEN)
MessageLoop(bot, handle).run_forever()
```
